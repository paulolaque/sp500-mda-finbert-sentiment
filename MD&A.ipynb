```python
# Listar nomes dos arquivos
import os
import glob
os.chdir(r"/content/data/mda")
filenames = [i for i in glob.glob("*.mda")]
```

## Colocar MD&A como uma coluna na tabela de empresas

```python
df['File Name']=df['File Name'].astype(str).str.replace('edgar/data/','')
df['File Name']=df['File Name'].map(lambda x: x.replace('/','_'))
df['File Name']=df['File Name'].map(lambda x: x.replace('.txt','.mda'))
```

```python
for fname in filenames:  
  with open(fname) as infile:
    df.loc[df['File Name']==fname, 'MD&A']=infile.read()
```
